# TASK 1(Data Cleaning and Preprocessing)

```python
##Importing Packages
import pandas as pd
```

```python
## Importing DataSet
SPD = pd.read_csv("2) Stock Prices Data Set.csv")
SPD
```

```python
## Checking for Missing Values
SPD.isnull().sum()
```

```python
SPD.sort_values(by=['symbol', 'date'])
```

```python
## Imputing missing values by combining forward and before data
SPD[['open', 'high', 'low']] = (
    SPD.groupby('symbol')[['open', 'high', 'low']]
    .ffill()
    .bfill()
)
```

```python
SPD.isnull().sum()
```

```python
## Checking for duplicate rows
SPD.duplicated().sum()
```

```python
## CHECKING FOR INCONSISTENT DATA FORMATS
## Understanding Data Types
SPD.dtypes
```

```python
## Standardizing the Date format
SPD.date.head()
```

```python
SPD.date = pd.to_datetime(SPD.date, errors='coerce')
SPD.date.head()
```
